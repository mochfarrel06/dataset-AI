```python
import cv2
import numpy as np
import imutils
from google.colab.patches import cv2_imshow
```

```python
# membaca file
image = cv2.imread('/gdrive/My Drive/kecerdasan buatan/sirih.jpg')
```

```python
# Menampilkan gambar
cv2_imshow(image)
```

```python
# Untuk mengetahui lebar dan tingginya
print(image.shape)
```

```python
# Resize image (mempertahankan efek rasio)
resized_1 = imutils.resize(image, width=500)
cv2_imshow(resized_1)
```

```python
# Save image
cv2.imwrite('/gdrive/My Drive/kecerdasan buatan/modifysirih.jpg', resized_1)
```

```python
# Resize image without aspect ratio
resized_2 = cv2.resize(image, (500,500))
cv2_imshow(resized_2)
```
